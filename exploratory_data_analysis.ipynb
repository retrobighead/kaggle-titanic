```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# データの読み込みと戦闘データの表示

data_path = './data/train.csv'
train_df = pd.read_csv(data_path, index_col=0)
train_df.head()
```

```python
# 各列の要約統計量を取得 @see(https://note.nkmk.me/python-pandas-describe/)
# 数値統計量のみの計算
train_df.describe()
```

```python
# カテゴリカル変数の要約統計量を算出
train_df.describe(include="object")
```

```python
# 欠損値の個数と型を確認
train_df.info()
```

```python
# 散布図行列
sns.pairplot(train_df)
```

```python
# 相関行列の確認 (数値変数)
cor = ｔrain_df.corr()
sns.heatmap(cor, square=True, cmap='plasma', vmax=1,vmin=-1, annot=True)
```
